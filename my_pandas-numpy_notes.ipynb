## Merge <a id='Merge'></a>

```python
#Now we'll learn about a many to many merge

# NOTE: many to many merge concept:
# i.e. two(2) Z in each dframe --> 
# relevant merging items would be (2x2 Z items) (in the current example)
# two(2) Z in df1 and one(1) Z in df2 --> 
# relevant merging items would be (2x1 Z items)
# three(3) Z in df1 and two(2) Z in df2 --> 
# relevant merging items would be (3x2 Z items)

df1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'dset1': np.arange(6)})
df2 = DataFrame({'key':['Q','Y','Z','Z'],'dset2':[1,2,3,4]})

df_inner = pd.merge(df1,df2, on='key', indicator='Ind.')
df_left = pd.merge(df1,df2, on='key', how='left', indicator='Ind.')
df_right = pd.merge(df1,df2, on='key', how='right', indicator='Ind.')
df_outer = pd.merge(df1,df2, on='key', how='outer', indicator='Ind.')

df_merge = pd.concat([df1,df2, df_inner, df_left, df_right, df_outer], axis=1, 
                    keys=['dframe1(df1)','dframe2(df2)', 'marge_inner (df1, df2)'
                           ,'marge_left (df1, df2)','marge_right (df1, df2)'
                           ,'marge_outer (df1, df2)'])

# Removing unnessary rows with NaN from each df
df_merge.loc[6:9,['dframe1(df1)','dframe2(df2)','marge_inner (df1, df2)']]='===='
df_merge.loc[4:6,['dframe2(df2)']] = '===='
df_merge.loc[8,['dframe1(df1)','dframe2(df2)','marge_inner (df1, df2)'
                ,'marge_left (df1, df2)','marge_right (df1, df2)']] = '===='
df_merge.loc[6:8,['marge_right (df1, df2)']] = '===='
df_merge
```

```python
# We can also merge with multiple keys!

# Dframe on left
df1 = DataFrame({'key1': ['SF', 'SF', 'LA'],'key2': ['one', 'two', 'one'],
                  'left_d': [10,20,30]})
#Dframe on right
df2 = DataFrame({'key1': ['SF', 'SF', 'LA', 'LA'],
                 'key2': ['one', 'one', 'one', 'two'],
                   'right_d': [40,50,60,70]})

#Merge
df_inner = pd.merge(df1, df2, on=['key1', 'key2'], how='inner', indicator='Ind.')
df_left = pd.merge(df1, df2, on=['key1', 'key2'], how='left', indicator='Ind.')
df_right = pd.merge(df1, df2, on=['key1', 'key2'], how='right', indicator='Ind.')
df_outer = pd.merge(df1, df2, on=['key1', 'key2'], how='outer', indicator='Ind.')

dframes = pd.concat([df1,df2], axis=1, 
                    keys=['left_dframe (df1)','right_dframe (df2)'])

df_merge = pd.concat([df_inner, df_left, df_right, df_outer], axis=1, 
                     keys=['marge_inner (df1, df2)','marge_left (df1, df2)'
                           ,'marge_right (df1, df2)','marge_outer (df1, df2)'] )

# Removing unnessary rows with NaN from each df
dframes.loc[3,'left_dframe (df1)'] = '===='
df_merge.loc[3:5,'marge_inner (df1, df2)'] = '===='
df_merge.loc[4,['marge_left (df1, df2)','marge_right (df1, df2)']] = '===='
dframes
df_merge
```

```python
# Note that the left and right DataFrames have overlapping key names (key1, key2).
# pandas automatically adds suffixes to them.
# Since there are common name for key1 and key2 in two dframe, 
# the key2 in two dframe will show as key2_x and key2_y.
pd.merge(df1, df2, on=['key1'])

# To more clarification which key2 belongs with which dataframe, 
# we can also specify what the suffix becomes
pd.merge(df1,df2, on='key1',suffixes=('_lefty','_righty'))
```

## Merge on Index <a id='Mergeind'></a>

```python
df_left = DataFrame({'key': ['X','Y','Z','X','Y'],
                  'data': range(5)}); df_left
df_right = DataFrame({'group_data': [10, 20]}, index=['X', 'Y']); df_right

#Now merge, we'll use the key for the left Dframe, and the index for the right
pd.merge(df_left,df_right,left_on='key',right_index=True)

# We can also get a union by using outer
pd.merge(df_left,df_right,left_on='key',right_index=True,how='outer')

# WE can also you .join()
df_left.join(df_right)
```

```python
#Now let's try something a little more complicated, remember hierarchal index?
df_left_hr = DataFrame({'key1': ['SF','SF','SF','LA','LA'],
                   'key2': [10, 20, 30, 20, 30],
                   'data_set': np.arange(5.)}); df_left_hr
df_right_hr = DataFrame(np.arange(10).reshape((5, 2)),
                   index=[['LA','LA','SF','SF','SF'],
                          [20, 10, 10, 10, 20]],
                   columns=['col_1', 'col_2']); df_right_hr

# Now we can merge the left by using keys and the right by its index
pd.merge(df_left_hr,df_right_hr,left_on=['key1','key2'],right_index=True)

# We can alo keep a union by choosing 'outer' method
pd.merge(df_left_hr,df_right_hr,left_on=['key1','key2'],
         right_index=True,how='outer')
```

## Pivoting <a id='Pivoting'></a>

```python
# Note: It is not necessary to understand how this dataset was made to 
# understand this Lecture.
# import pandas testing utility
import pandas.util.testing as tm; tm.N = 3

#Create a unpivoted function
def unpivot(frame):
    N, K = frame.shape
    
    data = {'value' : frame.values.ravel('F'),
            'variable' : np.asarray(frame.columns).repeat(N),
            'date' : np.tile(np.asarray(frame.index), K)}
    
    # Return the DataFrame
    return DataFrame(data, columns=['date', 'variable', 'value'])

#Set the DataFrame we'll be using
dframe = unpivot(tm.makeTimeDataFrame()); dframe

# First two values passed are the row and column indexes, then finally 
# an optional fill value
dframe_piv = dframe.pivot('date','variable','value'); dframe_piv
dframe_piv = dframe.pivot('variable','date','value'); dframe_piv
```

## GroupBy on DataFrames <a id='groupby'></a>

```python
dframe = DataFrame({'k1':['X','X','Y','Y','Z'],
                    'k2':['alpha','beta','alpha','beta','alpha'],
                    'dataset1':np.random.randn(5),
                    'dataset2':np.random.randn(5)}); dframe

# Now let's see how to use groupby
# Lets grab the dataset1 column and group it by the k1 key
group1 = dframe['dataset1'].groupby(dframe['k1']); group1
# Now we can perform operations on this particular group
group1.mean()
```

```python
# We can use group keys that are series as well
# We'll make some arrays for use as keys
cities = np.array(['NY','LA','LA','NY','NY'])
month = np.array(['JAN','FEB','JAN','FEB','JAN'])
# Now using the data from dataset1, group the means by city and month
dframe['dataset1'].groupby([cities,month]).mean()
```

```python
# --- Better and more general groupby commands ---
# WE can also pass column names as group keys 
dframe.groupby('k1').mean()
# Or multiple column names
dframe.groupby(['k1','k2']).mean()

# Next we'll learn how to use groupby with columns. For example if we only 
# wanted to group the dataset2 column with both sets of keys
dataset2_group = dframe.groupby(['k1', 'k2'])[['dataset2']]
dataset2_group.mean()

# Another useful groupby method is getting the group sizes
dframe.groupby(['k1']).size()
dframe.groupby(['k1','k2']).size()
```

## Aggregation <a id='agg'></a>

```python
# Data Agrregation consists of operations that result in a scalar 
# (e.g. mean(),sum(),count(), etc)
# Save thewinquality.csv file in the same folder as your ipython notebooks, 
# note the delimiter used ;
dframe_wine = pd.read_csv('winequality-red.csv',sep=';'); dframe_wine.head()

# How about we find out the average alcohol content for the wine
dframe_wine['alcohol'].mean()

# That was an example of an aggregate, how about we make our own?
def max_to_min(arr):
    return arr.max() - arr.min()

# Let's group the wines by "quality"
wino = dframe_wine.groupby('quality'); 
wino.describe()

# We can now apply our own aggregate function with "agg" syntax, this function 
# takes the max value of the col and subtracts the min value of the col.
wino.agg(max_to_min)

# We can also pass string methods through aggregate
wino.agg('mean')
```

```python
# Let's adda  quality to alcohol content ratio
dframe_wine['qual/alc ratio'] = dframe_wine['quality']/dframe_wine['alcohol']
dframe_wine.head()

# WE can also use pivot tables instead of groupby (Pivot table of quality)
dframe_wine.pivot_table(index=['quality'])
```

## Spilling, Applying and Combining <a id='spilling'></a>

```python
# Let's grab the wine data again
dframe_wine = pd.read_csv('winequality-red.csv',sep=';'); dframe_wine.head()

# What if we wanted to know the highest alcohol content for each quality range?
# We can use groupby mechanics to split-apply-combine.

# Create a function that assigns a rank to each wine based on alcohol content, 
# with 1 being the highest alcohol content
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df

# Now sort the dframe by alcohol in ascending order
dframe_wine.sort_values(by='alcohol',ascending=False,inplace=True)

# Now we'll group by quality and apply our ranking function
dframe_wine = dframe_wine.groupby('quality').apply(ranker); dframe_wine.head()

# Now finally we can just call for the dframe where the alc_content_rank == 1
# Get the numebr of quality counts
num_of_qual = dframe_wine['quality'].value_counts(); num_of_qual

# Now we'll show the combined info for teh wines that had the highest alcohol 
# content for their respective rank!
dframe_wine[dframe_wine.alc_content_rank == 1].head(len(num_of_qual))
```
